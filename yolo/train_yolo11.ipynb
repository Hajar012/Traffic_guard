**Deepsort**

```
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load your trained model
model = YOLO("runs/detect/train8/weights/best.pt")

# DeepSORT tracker
tracker = DeepSort(
    max_age=15,
    n_init=3,
    max_iou_distance=0.1,
    max_cosine_distance=0.3,
    nn_budget=100
)

# Adjust these based on your dataset
ACCIDENT_CLASS = 0
NONACCIDENT_CLASS = 1

track_labels = {}

cap = cv2.VideoCapture("test/carsAtDay.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, verbose=False)[0]

    detections = []
    temp_labels = []  # store labels for this frame

    for box in results.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])
        cls = int(box.cls[0])

        if conf < 0.5:
            continue

        # Only track cars
        if cls != NONACCIDENT_CLASS:
            continue

        w = x2 - x1
        h = y2 - y1
        xc = x1 + w/2
        yc = y1 + h/2

        detections.append(([xc, yc, w, h], conf, cls))

        # Determine accident label for this detection
        accident_label = "ACCIDENT" if cls == ACCIDENT_CLASS else "NON-ACCIDENT"
        temp_labels.append(accident_label)

    tracks = tracker.update_tracks(detections, frame=frame)

    for i, track in enumerate(tracks):
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        l, t, r, b = map(int, track.to_ltrb())

        # Assign label to this track
        if i < len(temp_labels):
            track_labels[track_id] = temp_labels[i]

        label = track_labels.get(track_id, "UNKNOWN")

        cv2.rectangle(frame, (l, t), (r, b), (0,255,0), 2)
        cv2.putText(frame, label, (l, t - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    cv2.imshow("Accident Detection + Tracking", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
