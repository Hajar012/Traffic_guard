```
import cv2
from ultralytics import YOLO

# Load models
accident_model = YOLO("runs/detect/train8/weights/best.pt")      # your trained model
coco_model = YOLO("yolo11n.pt")       # COCO pretrained model
cap = cv2.VideoCapture("test/Car accident Caught CCTV India _2.mp4")


def boxes_overlap(a, b):
    ax1, ay1, ax2, ay2 = a
    bx1, by1, bx2, by2 = b

    return not (ax2 < bx1 or ax1 > bx2 or ay2 < by1 or ay1 > by2)

while True:

    ret, frame = cap.read()
    if not ret:
        break

    # Run both models
    accident_results = accident_model(frame, conf=0.35)[0]
    coco_results = coco_model(frame, conf=0.35)[0]

    accident_boxes = []
    person_boxes = []

    # Collect accident model detections
    for box in accident_results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = float(box.conf[0])
        cls = int(box.cls[0])

        label = "accident" if cls == 0 else "non-accident"

        accident_boxes.append({
            "box": (x1, y1, x2, y2),
            "conf": conf,
            "label": label
        })

    # Collect person detections from COCO
    for box in coco_results.boxes:
        cls = int(box.cls[0])

        if cls == 0:  # COCO class 0 = person
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])

            person_boxes.append({
                "box": (x1, y1, x2, y2),
                "conf": conf,
                "label": "person"
            })

    drawn = []

    # Compare overlaps
    for a in accident_boxes:

        show_accident = True

        for p in person_boxes:

            if boxes_overlap(a["box"], p["box"]):

                # Compare confidence
                if p["conf"] > a["conf"]:

                    # draw person instead
                    x1, y1, x2, y2 = p["box"]

                    cv2.rectangle(frame, (x1,y1),(x2,y2),(0,255,0),2)
                    cv2.putText(frame,
                                f'person {p["conf"]:.2f}',
                                (x1, y1-10),
                                cv2.FONT_HERSHEY_SIMPLEX,
                                0.6,
                                (0,255,0),
                                2)

                    show_accident = False

                break

        if show_accident:

            x1, y1, x2, y2 = a["box"]

            color = (255,0,0) if a["label"]=="accident" else (0,255,255)

            cv2.rectangle(frame, (x1,y1),(x2,y2), color,2)
            cv2.putText(frame,
                        f'{a["label"]} {a["conf"]:.2f}',
                        (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.6,
                        color,
                        2)
    frame = cv2.resize(frame, (800, 450))
    cv2.imshow("YOLOv11 Detection", frame)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

**Deepsort**

```
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load your trained model
model = YOLO("runs/detect/train8/weights/best.pt")

# DeepSORT tracker
tracker = DeepSort(
    max_age=15,
    n_init=3,
    max_iou_distance=0.1,
    max_cosine_distance=0.3,
    nn_budget=100
)

# Adjust these based on your dataset
ACCIDENT_CLASS = 0
NONACCIDENT_CLASS = 1

track_labels = {}

cap = cv2.VideoCapture("test/carsAtDay.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, verbose=False)[0]

    detections = []
    temp_labels = []  # store labels for this frame

    for box in results.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])
        cls = int(box.cls[0])

        if conf < 0.5:
            continue

        # Only track cars
        if cls != NONACCIDENT_CLASS:
            continue

        w = x2 - x1
        h = y2 - y1
        xc = x1 + w/2
        yc = y1 + h/2

        detections.append(([xc, yc, w, h], conf, cls))

        # Determine accident label for this detection
        accident_label = "ACCIDENT" if cls == ACCIDENT_CLASS else "NON-ACCIDENT"
        temp_labels.append(accident_label)

    tracks = tracker.update_tracks(detections, frame=frame)

    for i, track in enumerate(tracks):
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        l, t, r, b = map(int, track.to_ltrb())

        # Assign label to this track
        if i < len(temp_labels):
            track_labels[track_id] = temp_labels[i]

        label = track_labels.get(track_id, "UNKNOWN")

        cv2.rectangle(frame, (l, t), (r, b), (0,255,0), 2)
        cv2.putText(frame, label, (l, t - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    cv2.imshow("Accident Detection + Tracking", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
